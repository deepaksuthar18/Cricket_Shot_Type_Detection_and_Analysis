# import all required libraries

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.regularizers import l2
import matplotlib.pyplot as plt
import os
import pickle
import numpy as np
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D
```

# set parameters for training

```python
IMG_SIZE = (128, 128)
BATCH_SIZE = 32
EPOCHS = 10
DATASET_PATH = r'C:\Users\deepa\Cricket Shot Type Detection and Analysis\Notebook and Dataset\Data'
```

# Load and Augment Image Data

```python
# Data Augmentation
datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

# Generate batches of augmented images
train_gen = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training'
)

val_gen = datagen.flow_from_directory(
    DATASET_PATH,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation'
)

print(f"Training images: {train_gen.samples}")
print(f"Validation images: {val_gen.samples}")
```

```python
import os
from collections import Counter

data_dir = r'C:\Users\deepa\Cricket Shot Type Detection and Analysis\Notebook and Dataset\Data'
class_counts = {folder: len(os.listdir(os.path.join(data_dir, folder))) for folder in os.listdir(data_dir)}
print("Class distribution:", class_counts)
```

# save label map


```python
label_map = train_gen.class_indices

# Save label map for future use
os.makedirs("model", exist_ok=True)
with open("model/label_map.pkl", "wb") as f:
    pickle.dump(label_map, f)

print("Class labels saved:", label_map)
```

# image with label

```python
# Get class label names
class_names = list(train_gen.class_indices.keys())

# Get a batch of images and labels from the generator
images, labels = next(train_gen)

# Plot first 9 images with labels
plt.figure(figsize=(10, 10))
for i in range(10):
    plt.subplot(5, 5, i + 1)
    plt.imshow(images[i])
    label_index = np.argmax(labels[i])  # Get the index of the hot-encoded label
    plt.title(class_names[label_index])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

# Load base model transfer learning

```python
base_model = MobileNetV2(input_shape=(*IMG_SIZE, 3), include_top=False, weights='imagenet')
base_model.trainable = False
```

```python
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)
x = Dense(64, activation='relu')(x)
output = Dense(train_gen.num_classes, activation='softmax')(x)

# Final model
model = Model(inputs=base_model.input, outputs=output)
```

```python
# Compile
model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
callbacks = [
    EarlyStopping(patience=3, restore_best_weights=True),
    ModelCheckpoint('model/best_model.h5', save_best_only=True)
]
```

# Train Model

```python
    history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=EPOCHS,
    callbacks=callbacks
)
```
